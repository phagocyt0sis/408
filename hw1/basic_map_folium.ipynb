```python
huntsville_location = (34.72806, -86.64733)
armor_location = (34.646166098419094, -86.77144334127473)
riverGuages = {
    "Limestone Creek" : (34.75166667, -86.82305556),
    "Big Wills Creek" : (34.43805556, -85.76722222),
    "Big Nance Creek" : (34.67, -87.31722222),
    "Flint River" : (34.74916667, -86.44666667),
    "Pinhook Creek" : (34.73027778, -86.59333333),
    "Tennessee River Decatur" : (34.61138889, -86.97527778),
    "Tennessee River Florence" : (34.78694444, -87.67),
    "Aldridge Creek Farley" : (34.62388889, -86.54111111),
    "Tennessee River below Guntersville Dam" : (34.42138889, -86.39388889),
    "Tennessee River above Guntersville Dam" : (34.4217, -86.3927),
    "Indian Creek" : (34.69722222, -86.69972222),
    "Bear Creek" : (34.44388889, -88.11555556),
    "Tennessee River Scottsboro" : (34.64361111, -85.98027778),
    "Aldridge Creek Sherwood Drive Huntsville" : (34.66416667, -86.55416667),
    "Aldridge Creek Toney Drive Huntsville" : (34.69805556, -86.55055556),
    "Paint Rock River" : (34.62416667, -86.30638889),
    "Tennessee River Whitesburg" : (34.57166667, -86.55805556)
}
m = folium.Map(location=huntsville_location, zoom_start=9, tiles="cartodb positron")
folium.Marker(
    location=armor_location,
    tooltip="Click me!",
    popup="ARMOR Radar",
    icon=folium.Icon(icon="cloud"),
).add_to(m)
for key, value in riverGuages.items():
    #print(value)
    folium.Marker(
        location = value,
        tooltip="Click me!",
        popup=key,
        icon=folium.Icon(color="green", icon="tint")
    ).add_to(m)

m
```
